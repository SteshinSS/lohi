```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cluster 1 (right) - 40 points
cluster1_x = np.random.normal(loc=0.5, scale=0.2, size=40)
cluster1_y = np.random.normal(loc=1.5, scale=0.2, size=40)

# Cluster 2 (left) - 10 points
cluster2_x = np.random.normal(loc=1.5, scale=0.2, size=10)
cluster2_y = np.random.normal(loc=0.5, scale=0.2, size=10)
```

```python
cluster1_x_shadow = cluster1_x + 0.003
cluster1_y_shadow = cluster1_y - 0.003
```

```python
#sns.set(style="white", rc={"figure.figsize":(8, 6)})
sns.set(style="white", rc={"figure.figsize":(6, 4)})
```

```python
plt.title('Hit Identification (Hi)')

# Remove the top and right spines for a cleaner look
#sns.despine()

# Customize the grid appearance
plt.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)

plt.scatter(cluster1_x, cluster1_y, s=80, facecolors='white', edgecolors='red', linewidths=1.5, label='Train')
plt.scatter(cluster2_x, cluster2_y, s=100, facecolors='white', edgecolors='blue', linewidths=1, marker='*' ,label='Test', alpha=0.7)
plt.legend()
plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])


plt.savefig('fig2.png', dpi=300, bbox_inches='tight')
plt.show()
```

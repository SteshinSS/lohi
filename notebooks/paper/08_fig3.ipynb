```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="white", rc={"figure.figsize":(6, 4)})
```

```python
cluster_scale = 0.1

train_x = list(np.random.normal(loc=0.5, scale=0.20, size=40))
train_y = list(np.random.normal(loc=0.5, scale=0.20, size=40))

cluster_1_x = np.random.normal(loc=0.7, scale=cluster_scale, size=6)
cluster_1_y = np.random.normal(loc=1.5, scale=cluster_scale, size=6)

cluster_2_x = np.random.normal(loc=2.3, scale=cluster_scale, size=6)
cluster_2_y = np.random.normal(loc=0.5, scale=cluster_scale, size=6)

cluster_3_x = np.random.normal(loc=1.8, scale=cluster_scale, size=6)
cluster_3_y = np.random.normal(loc=1.6, scale=cluster_scale, size=6)
```

```python
train_x.append(cluster_1_x[0])
cluster_1_x = cluster_1_x[1:]

train_y.append(cluster_1_y[0])
cluster_1_y = cluster_1_y[1:]

train_x.append(cluster_2_x[0])
cluster_2_x = cluster_2_x[1:]

train_y.append(cluster_2_y[0])
cluster_2_y = cluster_2_y[1:]

train_x.append(cluster_3_x[0])
cluster_3_x = cluster_3_x[1:]

train_y.append(cluster_3_y[0])
cluster_3_y = cluster_3_y[1:]
```

```python
def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke

    circle = Circle(
        (x, y),
        radius,
        clip_on=False,
        zorder=10,
        linewidth=1,
        edgecolor="black",
        facecolor=(0, 0, 0, 0.0125),
        path_effects=[withStroke(linewidth=5, foreground="w")],
    )
    ax.add_artist(circle)
```

```python
plt.title('Lead Optimization (Lo)')


ax = plt.gca()
circle(0.7, 1.5, 0.25)
circle(2.3, 0.5, 0.25)
circle(1.8, 1.6, 0.25)

# Customize the grid appearance
plt.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.5)

plt.scatter(cluster_1_x, cluster_1_y, s=100, facecolors='white', edgecolors='blue', linewidths=1, marker='*' ,label='Test', alpha=0.7)
plt.scatter(cluster_2_x, cluster_2_y, s=100, facecolors='white', edgecolors='blue', linewidths=1, marker='*' , alpha=0.7)
plt.scatter(cluster_3_x, cluster_3_y, s=100, facecolors='white', edgecolors='blue', linewidths=1, marker='*' ,alpha=0.7)
plt.scatter(train_x, train_y, s=80, facecolors='white', edgecolors='red', linewidths=1.5, label='Train')
#plt.legend()

handles, labels = plt.gca().get_legend_handles_labels()

#specify order of items in legend
order = [1, 0]

#add legend to plot
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order])

plt.gca().set_xticklabels([])
plt.gca().set_yticklabels([])
plt.xlim([0, 3])
plt.ylim([0, 2])



plt.savefig('fig3.png', dpi=300, bbox_inches='tight')
plt.show()
```

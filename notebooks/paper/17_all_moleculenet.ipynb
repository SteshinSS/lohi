# PDBBind

```python
import os
from rdkit import Chem
from rdkit.Chem import AllChem

# Function to convert an SDF file to a list of SMILES strings
def sdf_to_smiles(file_path):
    supplier = Chem.SDMolSupplier(file_path)
    smiles = []
    for mol in supplier:
        if mol is not None:
            smile = Chem.MolToSmiles(mol)
            smiles.append(smile)
    return smiles

# Your current directory
current_dir = '../../../v2015/'

# Empty list to store all SMILES strings
all_smiles = []

# Go through each folder in the current directory
for dir_name in os.listdir(current_dir):
    dir_path = os.path.join(current_dir, dir_name)
    # If it's a directory
    if os.path.isdir(dir_path):
        # Find the first .sdf file in this directory
        for file_name in os.listdir(dir_path):
            if file_name.endswith('.sdf'):
                file_path = os.path.join(dir_path, file_name)
                # Convert the SDF file to SMILES strings and add them to the list
                all_smiles.extend(sdf_to_smiles(file_path))
                # Since we only want one .sdf file per directory, we break after finding the first one
                break
```

```python
all_smiles
```

```python
good_smiles = []

for s in all_smiles:
    try:
        mol = Chem.MolFromSmiles(s)
        good_smiles.append(Chem.MolToSmiles(mol))
    except:
        pass
```

```python
len(good_smiles)
```

```python
idx = np.arange(len(good_smiles))
np.random.shuffle(idx)
all_smiles = np.array(good_smiles)
test_smiles = all_smiles[idx[:400]]
train_smiles = all_smiles[idx[400:]]
```

```python
test_smiles = list(test_smiles)
train_smiles = list(train_smiles)


nearest_dist, nearest_idx = get_nearest_mols(test_smiles, train_smiles, return_idx=True)

plt.hist(nearest_dist, bins=50)
plt.axvline(x=0.4, color = 'r', ls='--')
plt.title('PDBBind: The nearest distance between train and test')
plt.xlabel('ECFP4 Tanimoto Similarity')
```

```python
(np.array(nearest_dist) > 0.4).mean()
```

# PDBBind from DeepChem

```python
from deepchem.molnet.load_function.pdbbind_datasets import load_pdbbind 
from deepchem.feat.base_classes import DummyFeaturizer
feat = DummyFeaturizer()
dataset = load_pdbbind(featurizer=feat, set_name='refined')
```

```python
train_smiles = []

for files in dataset[1][0].X:
    sdf = files[0]
    supplier = Chem.SDMolSupplier(sdf)
    for mol in supplier:
        if mol is not None:
            try:
                # Convert the molecule object to a SMILES string
                smile = Chem.MolToSmiles(mol)
                m = Chem.MolFromSmiles(smile)
                smile = Chem.MolToSmiles(m)
                train_smiles.append(smile)
            except:
                pass

test_smiles = []

for files in dataset[1][2].X:
    sdf = files[0]
    supplier = Chem.SDMolSupplier(sdf)
    for mol in supplier:
        if mol is not None:
            try:
                # Convert the molecule object to a SMILES string
                smile = Chem.MolToSmiles(mol)
                m = Chem.MolFromSmiles(smile)
                smile = Chem.MolToSmiles(m)
                test_smiles.append(smile)
            except:
                pass
```

```python
nearest_dist, nearest_idx = get_nearest_mols(test_smiles, train_smiles, return_idx=True)

plt.hist(nearest_dist, bins=50)
plt.axvline(x=0.4, color = 'r', ls='--')
plt.title('PDBBind DeepChem: The nearest distance between train and test')
plt.xlabel('ECFP4 Tanimoto Similarity')
```

```python
(np.array(nearest_dist) > 0.4).mean()
```

```python
import pandas as pd
import numpy as np
from sklearn.dummy import DummyRegressor
```

```python
import sys
sys.path.append('../../../../code')

from metrics import get_lo_metrics
```

```python
def fit_predict(train, test):
    model = DummyRegressor()
    model.fit(None, train['value'])

    train_preds = model.predict(train)
    train_with_preds = train.copy()
    train_with_preds['preds'] = train_preds

    test_preds = model.predict(test)
    test_with_preds = test.copy()
    test_with_preds['preds'] = test_preds
    return train_with_preds, test_with_preds
```

```python
for i in [1, 2, 3]:
    train = pd.read_csv(f'../../../../data/lo/drd2/train_{i}.csv', index_col=0)
    test = pd.read_csv(f'../../../../data/lo/drd2/test_{i}.csv', index_col=0)
    train_with_preds, test_with_preds = fit_predict(train, test)
    
    print(f'Split {i}')
    print('Train metrics:')
    print(get_lo_metrics(train_with_preds, train_with_preds['preds']))
    print('Test metrics:')
    print(get_lo_metrics(test_with_preds, test_with_preds['preds']))
    print()

    train_with_preds.to_csv(f'../../../../predictions/lo/drd2/dummy/train_{i}.csv')
    test_with_preds.to_csv(f'../../../../predictions/lo/drd2/dummy/test_{i}.csv')
```
